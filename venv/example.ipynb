```python
import wfdb
import pandas as pd
from sktime.classification.distance_based import KNeighborsTimeSeriesClassifier

# Load the first record
record = wfdb.rdrecord('data/40689238')
df_record = pd.DataFrame(record.p_signal, columns=record.sig_name)

# Create a MultiIndex for X_train
instances = df_record.shape[1]  # Number of features (instances)
timepoints = df_record.shape[0]  # Number of time points

# Construct the MultiIndex: (instance, timepoint)
multi_index = pd.MultiIndex.from_product(
    [range(instances), range(timepoints)], names=["instance", "timepoint"]
)

# Create X_train with pd.MultiIndex specification
X_train = pd.DataFrame(
    df_record.values.flatten(), index=multi_index, columns=["value"]
)

# Create y_train as a 1D array-like object
y_train = pd.Series(record.sig_name, name="label")

# Load the second record
record = wfdb.rdrecord('data/41445586')
df_record = pd.DataFrame(record.p_signal, columns=record.sig_name)

# Construct the MultiIndex for X_test
instances_test = df_record.shape[1]  # Number of features (instances)
timepoints_test = df_record.shape[0]  # Number of time points

multi_index_test = pd.MultiIndex.from_product(
    [range(instances_test), range(timepoints_test)], names=["instance", "timepoint"]
)

# Create X_test with pd.MultiIndex specification
X_test = pd.DataFrame(
    df_record.values.flatten(), index=multi_index_test, columns=["value"]
)

# Create y_test as a 1D array-like object
y_test = pd.Series(record.sig_name, name="label")

# Initialize and fit the classifier
classifier = KNeighborsTimeSeriesClassifier(distance="euclidean")
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy = classifier.score(X_test, y_test)
```
